**a) Look for the missing values in all the columns and either impute them (replace with mean,
median, or mode) or drop them. Justify your action for this task.**

```python
missing_values = raw_data.isnull().sum()
print("Missing Values:")
print(missing_values)
```

Question b need to be addressed before completing question a

**b) Remove the units from some of the attributes and only keep the numerical values (for
example remove kmpl from “Mileage”, CC from “Engine”, bhp from “Power”, and lakh from
“New_price”).**

```python
raw_data['Mileage'] = pd.to_numeric(raw_data['Mileage'].str.replace(' km/kg| kmpl', '', regex=True), errors='coerce')
raw_data['Engine'] = pd.to_numeric(raw_data['Engine'].str.replace(' CC', '', regex=True), errors='coerce')
raw_data['Power'] = pd.to_numeric(raw_data['Power'].str.replace(' bhp', '', regex=True), errors='coerce')
raw_data['New_Price'] = pd.to_numeric(raw_data['New_Price'].str.replace(' Lakh', '', regex=True), errors='coerce')
```

```python
print("\nDataFrame after removing units:")
print(raw_data)
```

**Question A continuation**

```python
# Impute missing values
raw_data['Mileage'].fillna(raw_data['Mileage'].mean(), inplace=True)
raw_data['Engine'].fillna(raw_data['Engine'].mean(), inplace=True)
raw_data['Power'].fillna(raw_data['Power'].mean(), inplace=True)
raw_data['Seats'].fillna(raw_data['Seats'].mode()[0], inplace=True)
```

```python
raw_data.drop('New_Price', axis=1, inplace=True)
```

a. Justification:

I have used mean for numerical data and mode for categorical data when dealing with small number of missing values. The 'New_Price' column had a significant number of missing values (5032 out of the total rows), making imputation less reliable. Dropping it avoids potential bias.

```python
print("\nDataFrame after handling missing values:")
print(raw_data)
```

**C) Change the categorical variables (“Fuel_Type” and “Transmission”) into numerical one hot
encoded value**

```python
df = pd.get_dummies(raw_data, columns=['Fuel_Type', 'Transmission'], drop_first=True)
```

```python
print("\nDataFrame after one-hot encoding categorical variables:")
print(df)
```

**d) Create one more feature and add this column to the dataset**

```python
current_year = datetime.now().year
df['Current_Age'] = current_year - df['Year']
```

```python
print("\nDataFrame with the new 'Current_Age' feature:")
print(df)
```

**e) Perform select, filter, rename, mutate, arrange and summarize with group by operations (or
their equivalent operations in python) on this dataset.**

Select Operation:

```python
selected_columns = df[[ 'Owner_Type', 'Price']]
print("\nSelected Columns:")
print(selected_columns)
```

Filter Operation:

```python
filtered_data = df[df['Location'] == 'Chennai']
print("\nFiltered Data (Diesel cars only):")
print(filtered_data)
```

Rename Operation:

```python
df = df.rename(columns={'Kilometers_Driven': 'Kms_Driven'})
print("\nDataFrame with Renamed Columns:")
print(df)
```

Mutate operation:

```python
df['Power_Per_Seat'] = df['Power'] / df['Seats']
print("\nDataFrame with New 'Power_Per_Seat' Column:")
print(df)
```

Arrange operation:

```python
arranged_data = df.sort_values(by='Year', ascending=False)
print("\nDataFrame Sorted by 'Year' in Descending Order:")
print(arranged_data)
```

```python
summary_data = df.groupby('Fuel_Type_Electric')['Power'].mean().reset_index()
print("\nMean Price for Electric Fuel Type:")
print(summary_data)
```
